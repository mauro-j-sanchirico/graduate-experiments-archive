```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
session = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

```python
expr1 = wlexpr('Integrate[Cos[x*z]/(z^2 - a), {z, 0, c}]')
result = session.evaluate(expr1)

tex_integral = session.evaluate(wl.ToString(wl.TeXForm(result)))
display(Math(tex_integral))
```

```python
expr2 = wlexpr('Integrate[Cos[x*z]/(z^2 - a), {z, c, Infinity}]')
result = session.evaluate(expr2)

tex_integral = session.evaluate(wl.ToString(wl.TeXForm(result)))
display(Math(tex_integral))
```

```python
expr3 = wlexpr('Integrate[Cos[x*z]/(z^2 - a), {z, b, c}]')
result = session.evaluate(expr3)

tex_integral = session.evaluate(wl.ToString(wl.TeXForm(result)))
display(Math(tex_integral))
```

```python
session.stop()
```
