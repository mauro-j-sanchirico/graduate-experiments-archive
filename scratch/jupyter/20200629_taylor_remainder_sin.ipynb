```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
ws = WolframLanguageSession("C:\\Program Files\\Wolfram Research\\Wolfram Engine\\12.1\\WolframKernel.exe")

# Display Latex
from IPython.display import display, Math
```

## Taylor Series of sin(u)

```python
# Clear Variables
ws.evaluate('ClearAll[s]')
ws.evaluate('ClearAll[a]')

# Get the general coefficient formula using the Z-transform
ws.evaluate('a = SeriesCoefficient[Sin[u], {u, 0, 2m+1}]')

# Simplify the coefficient formula
ws.evaluate('a = Refine[a, {m >= 0, Element[m, Integers]}]')
ws.evaluate('a = Simplify[a]')
ws.evaluate('a = a /. I^(2m)->(-1)^m')

# Represent the Taylor Series as a Sum
ws.evaluate('s = Inactive[Sum][a u^(2m+1), {m, 0, M}]')

tex = ws.evaluate('ToString[TeXForm[Sin[u] == s]]')
display(Math(tex))
```

## Taylor Series of cos(u)

```python
# Clear Variables
ws.evaluate('ClearAll[s]')
ws.evaluate('ClearAll[a]')

# Get the general coefficient formula using the Z-transform
ws.evaluate('a = SeriesCoefficient[Cos[u], {u, 0, 2m}]')

# Simplify the coefficient formula
ws.evaluate('a = Refine[a, {m >= 0, Element[m, Integers]}]')
ws.evaluate('a = Simplify[a]')
ws.evaluate('a = a /. I^(2m)->(-1)^m')

# Represent the Taylor Series as a Sum
ws.evaluate('s = Inactive[Sum][a u^(2m), {m, 0, M}]')

tex = ws.evaluate('ToString[TeXForm[Cos[u] == s]]')
display(Math(tex))
```

## Partial Sum of sin(u)

```python
# Write the integral from the Taylor Remainder Theorem
ws.evaluate('R = Integrate[(u-t)^M Sin[Pi/2 (M+1) + t], t]')
ws.evaluate('R = Simplify[Refine[R, {M > 0, u > 0, t > 0, Element[M, Integers], Element[u, Reals]}]]')

tex = ws.evaluate('ToString[TeXForm[R]]')
display(Math(tex))

# Evaluate the integral at 0 and at u as per the theorem
ws.evaluate('R0 = R /. t->0')
ws.evaluate('R0 = Simplify[Refine[R0, {M > 0, u > 0, t > 0, Element[M, Integers], Element[u, Reals]}]]')
tex = ws.evaluate('ToString[TeXForm[R0]]')
display(Math(tex))

ws.evaluate('Ru = R /. t->u')
ws.evaluate('Ru = Simplify[Ru]')
tex = ws.evaluate('ToString[TeXForm[Ru]]')
display(Math(tex))

# Fundamental Theorem of Calculus
ws.evaluate('R0u = (Ru - R0)/Factorial[M]')
ws.evaluate('R0u = Simplify[R0u]')
ws.evaluate('R0u = Refine[Expand[R0u], {M >= 0, u > 0, t > 0, Element[M, Integers], Element[u, Reals]}]')
tex = ws.evaluate('ToString[TeXForm[R0u]]')
display(Math(tex))

# Check some values
ws.evaluate('R0uN = N[R0u /. {u->1, M->10}]')
tex = ws.evaluate('ToString[TeXForm[R0uN]]')
display(Math(tex))
```
