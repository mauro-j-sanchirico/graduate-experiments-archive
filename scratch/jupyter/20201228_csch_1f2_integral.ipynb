```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
session = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

```python
def calc(expr_str):
    expr = wlexpr(expr_str)
    result = session.evaluate(expr)
    tex_expr = session.evaluate(wl.ToString(wl.TeXForm(result)))
    display(Math(tex_expr))
```

```python
calc('MeijerGReduce[Sin[x], x]')
```

```python
calc('FullSimplify[Beta[Exp[pi*T], i*x/pi + 1/2, 0] -  Beta[Exp[pi*T], 1/2 - i*x/pi, 0]]')
```

```python
calc('Integrate[i/pi * u^(i*v/Pi - 1/2) / (1-u), u]')
```

```python
calc('FullSimplify[Integrate[i/pi * u^(i*v/Pi - 1/2) / (1-u), u]]')
```

```python
calc('Integrate[i/pi * u^(-1/2 - i*v/Pi) / (1-u), u]')
```
