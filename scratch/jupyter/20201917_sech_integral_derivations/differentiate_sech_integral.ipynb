```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
session = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

```python
def print_tex(expr):
    """Prints an expression in tex"""
    tex_expr = session.evaluate(wl.ToString(wl.TeXForm(expr)))
    display(Math(tex_expr))
```

## Compute an integral via Wolfram Language

```python
# Define the summand
session.evaluate('s1 := j! * x^(j-k+1) / (a^k)*(j-k+1) * (PolyLog[k, i*Exp[i*b-a*x]] - PolyLog[k,-i*Exp[i*b-a*x]])')
print_tex(session.evaluate('s1'))

# Divide by x^j
session.evaluate('s2 := s1/x^j')
print_tex(session.evaluate('s2'))

# Differentiate the summand
session.evaluate('s3 := D[s2, x]')
print_tex(session.evaluate('s3'))

# Simplify the derivative
session.evaluate('s4 := Simplify[s3]')
print_tex(session.evaluate('s4'))

# Sum the derivatives
session.evaluate('s5 := Simplify[Sum[s4, {k, 1, j+1}]]')
print_tex(session.evaluate('s5'))

session.evaluate('1+1 == 2')
```
