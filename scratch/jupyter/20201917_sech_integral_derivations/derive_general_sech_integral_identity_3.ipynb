# Derivation for a general integration identity involving sech

This notebook derives a general integration identity for functions involving sech and a power of a the variable of integration, i.e.

$$\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi} .$$

This notebook also serves as an example of how to easily interact with the Wolfram Language through Python.

```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
ws = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

## Helper functions

Here we define some helper functions to help print pretty outputs.

```python
def print_tex(expr):
    """Prints an expression in latex"""
    display(Math(expr))

def print_wlresult(expr):
    """Prints a wolfram language result in latex"""
    tex_expr = ws.evaluate(wl.ToString(wl.TeXForm(expr)))
    display(Math(tex_expr))

def print_wlexpr(expr):
    """Prints a wolfram language expression in latex"""
    tex_expr = ws.evaluate(wl.ToString(wl.TeXForm(ws.evaluate(expr))))
    display(Math(tex_expr))
    
def eval_wlprint(expr):
    """Evaluates an expression, stores, and prints the result"""
    result = ws.evaluate(expr)
    save_expr_str = 'rrr = {}'.format(expr)
    ws.evaluate(save_expr_str)
    print_wlresult(ws.evaluate('rrr'))

print('\nPrinting latex:')
print_tex('\sum_{i=1}^{10} x_i')

print('\nPrinting the result of an evaluated expression:')
print_wlresult(ws.evaluate('x+2+y+3'))

print('\nEvaluating and printing an expression directly:')
print_wlexpr('Sum[x^k, {k, 0, 5}]')

print('\nEvaluating, storing, and printing reults:')
eval_wlprint('Sum[x^k, {k, 0, 5}]')

print('\nAccessing those results in the variable "rrr":')
eval_wlprint('rrr*10 + y')
```

## Apply repeated integration by parts

Here we apply repeated integration by parts to our integral and use Wolfram Language to check the result at each step.

```python

print('The RIBP formula is given')

print_tex(r'''
\int \! \xi^j \mathrm{sech}(a \xi) \, \mathrm{d\xi}
    = \int \! f(\xi) \frac{\partial^n g(\xi)}{\partial \xi^n} \, \mathrm{d\xi}
    = (-1)^n \int \! \frac{\partial^n f(\xi)}{\partial \xi^n} g(\xi) \, \mathrm{d\xi}
    + \sum_{k=0}^{n-1} \!
        (-1)^k \frac{\partial^k f(\xi)}{\partial \xi^k}
               \frac{\partial^{n-k-1} g(\xi)}{\partial \xi^{n-k-1}}
''')

print('We choose:')

print_tex(r'''
\begin{align}
f(\xi) &= \xi^j \\
\frac{\partial^n g(\xi)}{\partial \xi^n} &= \mathrm{sech(a\xi)}\\
n &= j
\end{align}
''')

print('Representing the hyperbolic secant in terms of the polyogarithms to facilitate integration:')

print_tex(r'''
\begin{align}
\mathrm{sech(a\xi)} &= i \mathrm{Li}_0\left(-ie^{a\xi}\right) - i \mathrm{Li}_0\left(ie^{a\xi}\right) \\
        &= i \mathrm{Li}_0\left(-ie^{-a\xi}\right) - i \mathrm{Li}_0\left(ie^{-a\xi}\right)
\end{align}
''')
```

```python
print('We can derive the pieces of the IBP formula via repeated integration/differentiation:')

print_tex(r'''
\begin{align}
g(\xi) &= \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{j} \frac{\partial^j}{\partial z^j}  g(\xi) \\
&= \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{j} \, \mathrm{sech}(a\xi) \\
&= \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{j} \,
        i \left\{\mathrm{Li}_0\left(-i e^{-a\xi}\right) - \mathrm{Li}_0\left(i e^{-a\xi}\right)  \right\}
\end{align}
''')

print('Noting that:')

print_tex(r'''
\underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{n} \,
        \mathrm{Li}_m\left(i e^{-a\xi}\right) = \frac{(-1)^n}{a^n} \mathrm{Li}_{m+n}\left(-i e^{-a\xi}\right)
''')

print('Then:')

print_tex(r'''
\underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{n} \,
        \left\{\mathrm{Li}_0\left(-i e^{-a\xi}\right) - \mathrm{Li}_0\left(i e^{-a\xi}\right)  \right\} = 
        \frac{(-1)^n}{a^n} \left\{\mathrm{Li}_n\left(-i e^{-a\xi}\right) - \mathrm{Li}_n\left(i e^{-a\xi}\right)  \right\}
''')

print('Which implies:')

print_tex(r'''
\underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{j} \, \mathrm{sech}(a\xi) = 
        \frac{i(-1)^j}{a^j} \left\{\mathrm{Li}_j\left(-i e^{-a\xi}\right) - \mathrm{Li}_j\left(i e^{-a\xi}\right)  \right\}
''')

print('So then:')

print_tex(r'''
g(\xi) = \frac{i(-1)^j}{a^j}
        \left\{\mathrm{Li}_j\left(-i e^{-a\xi}\right) - \mathrm{Li}_j\left(i e^{-a\xi}\right)  \right\}
''')

print('The jth derivative of f:')

print_tex(r'''
\frac{\partial^j f(\xi)}{\partial \xi^j} = \frac{\partial^j \xi^j}{\partial \xi^j} = j!
''')

print('The kth derivative of f:')

print_tex(r'''
\frac{\partial^k f(\xi)}{\partial \xi^k} = \frac{\partial^k \xi^j}{\partial \xi^k} = \frac{j!}{(j-k)!}\xi^{j-k}
''')

print('The (j-k-1) derivative of g:')

print_tex(r'''
\begin{align}
\frac{\partial^{j-k-1} g(\xi)}{\partial \xi^{j-k-1}} 
&= \frac{\partial^{j-k-1}}{\partial \xi^{j-k-1}}
      \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{j}
      \frac{i(-1)^j}{a^j} \left\{\mathrm{Li}_j\left(-i e^{-a\xi}\right) - \mathrm{Li}_j\left(i e^{-a\xi}\right) \right\} \\
&= \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! \mathrm{d\xi}}_{k + 1}
      \frac{i(-1)^j}{a^j} \left\{\mathrm{Li}_j\left(-i e^{-a\xi}\right) - \mathrm{Li}_j\left(i e^{-a\xi}\right) \right\} \\
&= \frac{i(-1)^{j+k+1}}{a^{j+k+1}} \left\{\mathrm{Li}_{j+k+1}\left(-i e^{-a\xi}\right)
        - \mathrm{Li}_{j+k+1}\left(i e^{-a\xi}\right)  \right\}
\end{align}
''')
```

```python
print('Substituting back into the RIBP formula:')

print_tex(r'''
\begin{align}
\int \! \xi^j \mathrm{sech}(a\xi) \, \mathrm{d\xi}
   &= \int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \mathrm{d \xi} \\
   &= (-1)^j \int \! \frac{\partial^j f(\xi)}{\partial \xi^j} g(\xi) \mathrm{d\xi}
           + \sum_{k=0}^{j-1} (-1)^k
                   \frac{\partial^k f(\xi)}{\partial \xi^k}
                   \frac{\partial^{j-k-1} g(\xi)}{\partial \xi^{j-k-1} } \\
   &= (-1)^j \int \!
           j! \frac{i(-1)^j}{a^j} \left\{
               \mathrm{Li}_j\left(-i e^{a\xi}\right) - \mathrm{Li}_j\left(i e^{a\xi}\right)  \right\}
           \mathrm{d\xi}
           + \sum_{k=0}^{j-1} (-1)^k
                   \frac{j!}{(j-k)!}\xi^{j-k}
                   \frac{i(-1)^{j+k+1}}{a^{j+k+1}} \left\{\mathrm{Li}_{j+k+1}\left(-i e^{-a\xi}\right)
                       - \mathrm{Li}_{j+k+1}\left(i e^{-a\xi}\right)  \right\}
\end{align}
''')
```

```python
print('Checking the substitutions:')

wl.Clear("Global`*")

print_wlexpr('j = 4')
print_wlexpr('f = x^j')
print_wlexpr('g = I / a^j * (PolyLog[j, -I*Exp[-a*x]] - PolyLog[j, I*Exp[-a*x]])')

print_wlexpr('''
L = FullSimplify[Integrate[x^j * Sech[a*x], x]]
''')

print_wlexpr('''
R = FullSimplify[(-1)^j * Integrate[ D[f, {x, j}]*g, x] + Sum[(-1)^k * D[f, {x, k}] * D[g, {x, j-k-1}], {k, 0, j-1}]]
''')

# Check if the integral is equalivalent to the summation
print_wlexpr('FullSimplify[L == R]')
```
