# Derivation for a general integration identity involving sech

This notebook derives a general integration identity for functions involving sech and a power of a the variable of integration, i.e.

$$\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi} .$$

The indefinite integration formula is derived through two successive applications of repeated integration by parts and the use of generalized repeated integrals and derivatives.

This notebook also serves as an example of how to easily interact with the Wolfram Language through Python.

```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
ws = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

## Helpful formulas

Here we list some helpful formulas for refference.

```python
print('Repeated integration by parts:')

print_tex(r'''
\int \! f(z) \frac{\partial^n g(z)}{\partial z^n} \, \mathrm{dz}

    = (-1)^n \int \! \frac{\partial^n f(z)}{\partial z^n} g(z) \, \mathrm{dz}
    
    + \sum_{k=0}^{n-1} \!
        (-1)^k \frac{\partial^k f(z)}{\partial z^k}
               \frac{\partial^{n-k-1} g(z)}{\partial z^{n-k-1}}
''')

print('Repeated differentiation and integration involving polylogarithms and exponential:')
print_tex(r'''
\frac{\partial^n}{\partial z^n} \mathrm{Li}_m (ie^{az}) =

    a^n \mathrm{Li}_{m-n}(ie^{az})
''')
print_tex(r'''
\underbrace{\int \! \mathrm{dz} \, \cdots \int \! dz}_{n} \, \mathrm{Li}_m(ie^{az}) =

    \frac{1}{a^n}\mathrm{Li}_{m+n}(ie^{az})
''')

print('Repeated differentiation and integration of polynomials:')

print_tex(r'''
\frac{\partial^n}{\partial z^n} z^m = \frac{m!}{(m-n)!} z^{n-k}
''')

print_tex(r'''
\frac{\partial^n}{\partial z^n} z^n = n!
''')

print_tex(r'''
\underbrace{\int \! \mathrm{dz} \, \cdots \int \! dz}_{n} \, z^m =

    \frac{m!}{(m+n)!}z^{m+n}
''')

print_tex(r'''
\underbrace{\int \! \mathrm{dz} \, \cdots \int \! dz}_{n} \, z^n =

    \frac{n!}{(2n)!}z^{2n}
''')

print('Repeated differentiation of the secant:')
print_tex(r'''
\frac{\partial^n}{\partial z^n} \mathrm{sech}(az) = 

ia^n\left(\mathrm{Li}_{-n}(-ie^{az}) - \mathrm{Li}_{-n}(ie^{az})\right)
''')
```

## Apply repeated integration by parts

Here we apply repeated integration by parts to our integral.

```python
print_tex(r'''
\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi} = 

    \int \! f(\xi) \frac{\partial^n g(\xi)}{\partial \xi^n} \, \mathrm{d\xi}
''')

print_tex(r'''
f(\xi) = \mathrm{sech(a\xi)}
''')

print_tex(r'''
\frac{\partial^n g(\xi)}{\partial \xi^n} = \xi^j
''')
```

## Derive the pieces of the integration by parts formula

Here we derive the pieces of the integration by parts formula to start reducing the indefinite integral.

```python
print('To generate:')
print_tex(r'''
\int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = (-1)^j \int \! \frac{\partial^j f(\xi)}{\partial \xi^j} g(\xi) \, \mathrm{d\xi}
    
    + \sum_{k=0}^{j-1} \!
        (-1)^k \frac{\partial^k f(\xi)}{\partial \xi^k}
               \frac{\partial^{j-k-1} g(\xi)}{\partial \xi^{j-k-1}}
''')

print('We need to produce:')
print_tex(r'''
g(\xi) = \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{j} \,

    \frac{\partial^j g(\xi)}{\partial \xi^j} =
    
    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{j} \, \xi^j = 
    
    \frac{j!}{\left(2j\right)!}\xi^{2j}
''')

print_tex(r'''
\frac{\partial^j f(\xi)}{\partial \xi^j} =

    \frac{\partial^j}{\partial \xi^j} \mathrm{sech(a\xi)} = 
    
    ia^j\left(\mathrm{Li}_{-j}(-ie^{az}) - \mathrm{Li}_{-j}(ie^{az})\right)
''')

print_tex(r'''
\frac{\partial^{j-k-1} g(\xi)}{\partial \xi^{j-k-1}} = 

    \frac{\partial^{j-k-1}}{\partial \xi^{j-k-1}}

    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{j} \, \xi^j = 
    
    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{k+1} \, \xi^j = 
    
    \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')
```

```python
print('Substituting:')
print_tex(r'''
\int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = (-1)^j \int \!
        \frac{\partial^j f(\xi)}{\partial \xi^j}
        g(\xi) \,
    \mathrm{d\xi}
    
    + \sum_{k=0}^{j-1} \!
    
        (-1)^k \frac{\partial^k f(\xi)}{\partial \xi^k}
        
               \frac{\partial^{j-k-1} g(\xi)}{\partial \xi^{j-k-1}}
''')

print_tex(r'''
\int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = (-1)^j \int \!
        ia^j\left(\mathrm{Li}_{-j}(-ie^{az}) - \mathrm{Li}_{-j}(ie^{az})\right)
        \frac{j!}{\left(2j\right)!}\xi^{2j} \,
    \mathrm{d\xi}
    
    + \sum_{k=0}^{j-1} \!
    
        (-1)^k ia^k\left(\mathrm{Li}_{-k}(-ie^{az}) - \mathrm{Li}_{-k}(ie^{az})\right)
        
               \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')
```

```python
print('Simplifying:')

print_tex(r'''
\int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = i(-1)^j a^j \frac{j!}{\left(2j\right)!} \int \!
        \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \,
    \mathrm{d\xi}
    - \int \!
        \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \,
    \mathrm{d\xi}   
    
    + \sum_{k=0}^{j-1} \!
    
        (-1)^k ia^k\left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
        
               \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')
```

## Use a second repeated integration by parts to handle the polylogarithms

Here we apply integration by parts a second time to handle the polylogarithm terms:

$$\int \! \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \, \mathrm{d\xi}$$

We seek a general integration identity for:

$$\int \! \xi^{n}\mathrm{Li}_{-m}(ie^{a\xi}) \, \mathrm{d\xi}$$

```python
print('To generate:')
print_tex(r'''
\int \! \xi^{n}\mathrm{Li}_{-m}(ie^{a\xi}) \, \mathrm{d\xi}

    = \int \! f(\xi) \frac{\partial^n g(\xi)}{\partial \xi^n} \, \mathrm{d\xi}

    = (-1)^n \int \! \frac{\partial^n f(\xi)}{\partial \xi^n} g(\xi) \, \mathrm{d\xi}
    
    + \sum_{k=0}^{n-1} \!
        (-1)^k \frac{\partial^k f(\xi)}{\partial \xi^k}
               \frac{\partial^{n-k-1} g(\xi)}{\partial \xi^{n-k-1}}
''')

print('We need to produce:')

print_tex(r'''
g(\xi) = \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{n} \,

    \frac{\partial^n g(\xi)}{\partial \xi^n} =
    
    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{n} \, \mathrm{Li}_{-m}(ie^{a\xi}) =
    
    \frac{1}{a^{n}}\mathrm{Li}_{n - m}(ie^{a\xi})
''')

print_tex(r'''
\frac{\partial^n f(\xi)}{\partial \xi^n} =

    \frac{\partial^n}{\partial \xi^n} \xi^n = n!
''')

print_tex(r'''
\frac{\partial^k f(\xi)}{\partial \xi^k} =

    \frac{\partial^k}{\partial \xi^k} \xi^n =
    
    \frac{n!}{(n-k)!} \xi^{n-k}
''')

print_tex(r'''
\frac{\partial^{n-k-1} g(\xi)}{\partial \xi^{n-k-1}} = 

    \frac{\partial^{n-k-1}}{\partial \xi^{n-k-1}}

    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{n} \, \mathrm{Li}_{-m}(ie^{a\xi})= 
    
    \underbrace{\int \! \mathrm{d\xi} \, \cdots \int \! d\xi}_{k+1} \, \mathrm{Li}_{-m}(ie^{a\xi}) = 
    
    \frac{1}{a^{k+1}} \mathrm{Li}_{k+1-m}(ie^{a\xi})
''')
```

```python
print('Substituting:')

print_tex(r'''
\int \! \xi^{n}\mathrm{Li}_{-m}(ie^{a\xi}) \, \mathrm{d\xi}

    = (-1)^n \int \!
    
        n! \frac{1}{a^{n}}\mathrm{Li}_{n - m}(ie^{a\xi}) \,
        
    \mathrm{d\xi}
    
    + \sum_{k=0}^{n-1} \!
    
        (-1)^k
        
        \frac{n!}{(n-k)!} \xi^{n-k}
        
        \frac{1}{a^{k+1}} \mathrm{Li}_{k+1-m}(ie^{a\xi})
''')
```

```python
print('Simplifying:')

print_tex(r'''
\int \! \xi^{n}\mathrm{Li}_{-m}(ie^{a\xi}) \, \mathrm{d\xi}

    =  \frac{(-1)^n n!}{a^{n}} \int \!
    
        \mathrm{Li}_{n - m}(ie^{a\xi}) \,
        
    \mathrm{d\xi}
    
    + \sum_{k=0}^{n-1} \!
    
        \frac{(-1)^k n! \xi^{n-k}}{(n-k)! a^{k+1}} \mathrm{Li}_{k+1-m}(ie^{a\xi})
''')
```

```python
print('Integrating:')

print_tex(r'''
\int \! \xi^{n}\mathrm{Li}_{-m}(ie^{a\xi}) \, \mathrm{d\xi}

    =  \frac{(-1)^n n!}{a^{n+1}} \mathrm{Li}_{n - m + 1}(ie^{a\xi}) \,
    
    + \sum_{k=0}^{n-1} \!
    
        \frac{(-1)^k n! \xi^{n-k}}{(n-k)! a^{k+1}} \mathrm{Li}_{k+1-m}(ie^{a\xi})
''')

print('Due to symmetry:')
print_tex(r'''
\int \! \xi^{n}\mathrm{Li}_{-m}(-ie^{a\xi}) \, \mathrm{d\xi}

    =  \frac{(-1)^n n!}{a^{n+1}} \mathrm{Li}_{n - m + 1}(-ie^{a\xi}) \,
    
    + \sum_{k=0}^{n-1} \!
    
        \frac{(-1)^k n! \xi^{n-k}}{(n-k)! a^{k+1}} \mathrm{Li}_{k+1-m}(-ie^{a\xi})
''')
```

```python
print('Returning to the original function:')
print_tex(r'''
\int \! \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \, \mathrm{d\xi}

=  \frac{(2j)!}{a^{2j+1}} \mathrm{Li}_{j + 1}(ie^{a\xi}) \,
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \mathrm{Li}_{k+1-j}(ie^{a\xi})
''')

print_tex(r'''
\int \! \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \, \mathrm{d\xi}

=  \frac{(2j)!}{a^{2j+1}} \mathrm{Li}_{j + 1}(-ie^{a\xi}) \,
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \mathrm{Li}_{k+1-j}(-ie^{a\xi})
''')
```

## Use Wolfram Language to put the pieces together

Here we use the Wolfram language to substitute and simplify the integration by parts formula.  We return to the intermediate result:

```python
print_tex(r'''

\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi}

    = \int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = i(-1)^j a^j \frac{j!}{\left(2j\right)!} \left(
    
    \int \!
        \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \,
    \mathrm{d\xi}
    
    - \int \!
        \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \,
    \mathrm{d\xi}   
    \right)
    
    + \sum_{k=0}^{j-1} \!
    
        (-1)^k ia^k\left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
        
               \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')
```

```python
print('Noting:')
print_tex(r'''
\int \! \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \, \mathrm{d\xi}

=  \frac{(2j)!}{a^{2j+1}} \mathrm{Li}_{j + 1}(ie^{a\xi}) \,
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \mathrm{Li}_{k+1-j}(ie^{a\xi})
''')

print_tex(r'''
\int \! \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \, \mathrm{d\xi}

=  \frac{(2j)!}{a^{2j+1}} \mathrm{Li}_{j + 1}(-ie^{a\xi}) \,
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \mathrm{Li}_{k+1-j}(-ie^{a\xi})
''')

print('Then:')
print_tex(r'''
    \int \!
        \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \,
    \mathrm{d\xi}
    
    - \int \!
        \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \,
    \mathrm{d\xi}
    
    = \frac{(2j)!}{a^{2j+1}} \left(
        \mathrm{Li}_{j + 1}(-ie^{a\xi})  - \mathrm{Li}_{j + 1}(ie^{a\xi}) 
    \right)
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \left(
        \mathrm{Li}_{k+1-j}(-ie^{a\xi}) - \mathrm{Li}_{k+1-j}(ie^{a\xi})
    \right)
    
''')

print_tex(r'''
    \int \!
        \xi^{2j}\left(\mathrm{Li}_{-j}(-ie^{a\xi}) - \mathrm{Li}_{-j}(ie^{a\xi}) \right) \,
    \mathrm{d\xi}
    
    = \frac{(2j)!}{a^{2j+1}} \left(
        \mathrm{Li}_{j + 1}(-ie^{a\xi})  - \mathrm{Li}_{j + 1}(ie^{a\xi}) 
    \right)
    
    + \sum_{k=0}^{2j-1} \!
    
    \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \left(
        \mathrm{Li}_{k+1-j}(-ie^{a\xi}) - \mathrm{Li}_{k+1-j}(ie^{a\xi})
    \right)
    
''')
```

```python
print('testing:')

wl.Clear("Global`*")

print_wlexpr('j = 2')

print_wlexpr('''
q = (2*j)! / a^(2*j + 1) * (PolyLog[j+1, -I*Exp[a*x]] - PolyLog[j+1, I*Exp[a*x]]) \
    + Sum[ \
        (-1)^k*(2*j)!*x^(2*j-k) / ((2*j-k)! * a^(k+1)) * (PolyLog[k+1-j, -I*Exp[a*x]]- PolyLog[k+1-j, I*Exp[a*x]]), \
        {k, 0, 2*j - 1}]
''')

print_wlexpr('''
r = Integrate[x^(2*j)*(PolyLog[-j, -I*Exp[a*x]] - PolyLog[-j, I*Exp[a*x]]), x]
''')

# Check if the integral is equalivalent to the summation
print_wlexpr('FullSimplify[r == q]')
```

```python
print('Returing again to the intermediate result:')

print_tex(r'''

\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi}

    = \int \! f(\xi) \frac{\partial^j g(\xi)}{\partial \xi^j} \, \mathrm{d\xi}

    = i(-1)^j a^j \frac{j!}{\left(2j\right)!} \left(
    
    \int \!
        \xi^{2j}\mathrm{Li}_{-j}(-ie^{a\xi}) \,
    \mathrm{d\xi}
    
    - \int \!
        \xi^{2j}\mathrm{Li}_{-j}(ie^{a\xi}) \,
    \mathrm{d\xi}   
    \right)
    
    + \sum_{k=0}^{j-1} \!
    
        (-1)^k ia^k\left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
        
               \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')

print('Substituting in the polylogarithm integrals:')

print_tex(r'''
\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi}

= i(-1)^j a^j \frac{j!}{\left(2j\right)!} \left(
    
        \frac{(2j)!}{a^{2j+1}} \left(
            \mathrm{Li}_{j + 1}(-ie^{a\xi})  - \mathrm{Li}_{j + 1}(ie^{a\xi}) 
        \right)
    
        + \sum_{k=0}^{2j-1} \!
    
        \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \left(
            \mathrm{Li}_{k+1-j}(-ie^{a\xi}) - \mathrm{Li}_{k+1-j}(ie^{a\xi})
        \right)
   
    \right)
''')

print_tex(r'''
    + \sum_{k=0}^{j-1} \! (-1)^k ia^k\left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
        
               \frac{j!}{(j+k+1)!} \xi^{j+k+1}
''')

print('Simplifying:')

print_tex(r'''
\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi}

   = i(-1)^j \frac{j!}{a^{j+1}} \left(
    
        \left( \mathrm{Li}_{j + 1}(-ie^{a\xi})  - \mathrm{Li}_{j + 1}(ie^{a\xi}) \right)
    
        + \sum_{k=0}^{2j-1} \!
    
        \frac{(-1)^k (2j)! \xi^{2j-k}}{(2j-k)! a^{k+1}} \left(
            \mathrm{Li}_{k+1-j}(-ie^{a\xi}) - \mathrm{Li}_{k+1-j}(ie^{a\xi})
        \right)
   
    \right)
''')

print_tex(r'''
    + \sum_{k=0}^{j-1} \! \left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
        
               \frac{(-1)^k ia^k j!}{(j+k+1)!} \xi^{j+k+1}
''')

print('Simplifying:')

print_tex(r'''
\int \! \xi^j \mathrm{sech(a\xi)} \, \mathrm{d\xi}

   =  \frac{i (-1)^j j!}{a^{j+1}} \left( \mathrm{Li}_{j + 1}(-ie^{a\xi})  - \mathrm{Li}_{j + 1}(ie^{a\xi}) \right)
    
        + i \sum_{k=0}^{2j-1} \!
    
            \frac{(-1)^{j+k} j! (2j)! \xi^{2j-k}}{(2j-k)! a^{j+k+2}} \left(
                \mathrm{Li}_{k+1-j}(-ie^{a\xi}) - \mathrm{Li}_{k+1-j}(ie^{a\xi})
            \right)
''')

print_tex(r'''
    + i \sum_{k=0}^{j-1} \! \frac{(-1)^k a^k j! \xi^{j+k+1}}{(j+k+1)!} 
        
        \left(\mathrm{Li}_{-k}(-ie^{a\xi}) - \mathrm{Li}_{-k}(ie^{a\xi})\right)
''')
```

```python
print('testing:')

wl.Clear("Global`*")

print_wlexpr('j = 1')

print_wlexpr('''
q1 = I*(-1)^j * j! / a^(j+1) * (PolyLog[j+1, -I*Exp[a*x]] - PolyLog[j+1, I*Exp[a*x]]);
q2 = I*Sum[(-1)^(j+k) * j! * (2*j)! * x^(2*j-k) / (((2*j - k)! * a^(j+k+2)) * (PolyLog[k+1-j, -I*Exp[a*x]] - PolyLog[k+1-j, I*Exp[a*x]])), {k, 0, 2*j-1}];
q3 = I*Sum[(-1)^k * a^k * j! * x^(j+k+1) / (j+k+1)! * (PolyLog[-k, -I*Exp[a*x]] - PolyLog[-k, I*Exp[a*x]]), {k, 0, j-1}];

q = q1 + q2 + q3
''')

print_wlexpr('''
r = Integrate[x^j * Sech[a*x], x]
''')

# Check if the integral is equalivalent to the summation
#print_wlexpr('FullSimplify[r - q]')
```
