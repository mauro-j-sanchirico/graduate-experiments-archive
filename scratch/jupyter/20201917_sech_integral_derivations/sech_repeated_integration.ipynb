```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
session = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

## Compute an integral via Wolfram Language

```python
x_j_integrand = wlexpr('x^j')
sech_integrand = wlexpr('Sech[a*x]')
intvar = wlexpr('x')

x_j_integral = session.evaluate(wl.Integrate(x_j_integrand, intvar))
x_j_tex_integral = session.evaluate(wl.ToString(wl.TeXForm(x_j_integral)))
display(Math(x_j_tex_integral))

sech_integral = session.evaluate(wl.Integrate(sech_integrand, intvar))
sech_tex_integral = session.evaluate(wl.ToString(wl.TeXForm(sech_integral)))
display(Math(sech_tex_integral))
```

## Define a function for repeated integration

```python
def compute_repeated_integral(integrand, intvar, num_integrations):
    """Computes a repeated integral"""
    
    integral = session.evaluate(wl.Integrate(integrand, intvar))
    
    for idx in range(num_integrations-1):
        integral = session.evaluate(wl.Integrate(integral, intvar))
        
    tex_integral = session.evaluate(wl.ToString(wl.TeXForm(wl.Simplify(integral))))
    
    return (integral, tex_integral)
```

## Compute the repeated integrals

```python
max_num_integrations = 10

for num_integrations in range(1, max_num_integrations+1):
    integral, tex_integral = compute_repeated_integral(x_j_integrand, intvar, num_integrations)
    display(Math(tex_integral))
```

```python
max_num_integrations = 4

for num_integrations in range(1, max_num_integrations+1):
    integral, tex_integral = compute_repeated_integral(sech_integrand, intvar, num_integrations)
    display(Math(tex_integral))
```
