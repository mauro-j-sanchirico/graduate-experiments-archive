# Polynomial Math Examples

This notebook provides an example of how to use the polynomialmath library.  The polynomialmath library is a custom developed library for performing basic arithmetic operations on polynomials.  It is meant to serve as a foundation for formal verifed computing with polynomials via
Taylor models and Taylor-like models.  This notebook shows how to use the basic polynomial arithmetic functions provided by polynomialmath.


```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from verified_computing import polynomialmath
from verified_computing import intvmath
```

## Examples of how to use the basic operations

Basic operations supported:
* Polynomial addition
* Polynomial subtraction
* Polynomial multiplication
* Polynomial evaluation
* Polynomial bounding


## Printing

```python
print('Polynomial printing examples:')
p1 = polynomialmath.Polynomial(np.array([-10.0, -3.2, 2, -5, 7]))
p2 = polynomialmath.Polynomial(np.array([2, 0, -5, 7]))
p3 = polynomialmath.Polynomial(np.array([0, 0, 1, 0]))
p4 = polynomialmath.Polynomial(np.array([0, -1, 0, 0]))
p5 = polynomialmath.Polynomial(np.array([10, 0, -2, 0]))
p6 = polynomialmath.Polynomial(np.array([1]))
p7 = polynomialmath.Polynomial(np.array([0]))

p1.print()
p2.print()
p3.print()
p4.print()
p5.print()
p6.print()
p7.print()

print()
```

## Copying a polynomial

Polynomials need to be copied using the .copy() function to allow array members storing coefficients and exponents to be copied by value and not by refference.


```python
pa = polynomialmath.Polynomial(np.array([1,1,1,1]))
pb = pa.copy()
pc = pa

pb.coef_vector[-1] = 10

print('pa does not change: {}'.format(pa.tostr()))

pc.coef_vector[-1] = 10

print('Now pa does change: {}'.format(pa.tostr()))
```

## Addition

Polynomials can be added together.


```python
print('Polynomial addition example 1:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2, 2, 2, 2]))
pr = pa + pb

print('({})  +  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial addition example 2:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1, 1, 1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2]))
pr = pa + pb

print('({})  +  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial addition example 3:')
pa = polynomialmath.Polynomial(np.array([1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2]))
pr = pa + pb

print('({})  +  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Add a scalar

Scalars can be added to polynomials.


```python
print('Polynomial scalar addition example:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
scalar = 10
pr = polynomialmath.add_scalar_to_polynomial(scalar, pa)

print('{}  +  ({})'.format(scalar, pa.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Subtraction

One polynomial can be subtracted from another.


```python
print('Polynomial subtraction example 1:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2, 2, 2, 2]))
pr = pa - pb

print('({})  -  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial subtraction example 2:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
pb = polynomialmath.Polynomial(np.array([-2, -2, -2, -2, -2]))
pr = pa - pb

print('({})  -  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial subtraction example 2:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
pb = polynomialmath.Polynomial(np.array([-2, -2, -2, -2, -2]))
pr = pa - pb

print('({})  -  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Multiplication

Two polynomials can be multiplied together.



```python
print('Polynomial multiplication example 1:')
pa = polynomialmath.Polynomial(np.array([1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2]))
pr = pa * pb

print('({})  *  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial multiplication example 2:')
pa = polynomialmath.Polynomial(np.array([-1]))
pb = polynomialmath.Polynomial(np.array([-10, 2, 3, -4]))
pr = pa * pb

print('({})  *  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial multiplication example 3:')
pa = polynomialmath.Polynomial(np.array([-10, 4, 5, -2]))
pb = polynomialmath.Polynomial(np.array([-2, -4]))
pr = pa * pb

print('({})  *  ({})'.format(pa.tostr(), pb.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Multiply a scalar

Polynomials can be multiplied by a scalar.


```python
print('Polynomial scalar multiplication example:')
pa = polynomialmath.Polynomial(np.array([1, 1, 1]))
scalar = 10
pr = polynomialmath.multiply_polynomial_by_scalar(scalar, pa)

print('{}  *  ({})'.format(scalar, pa.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Approximate Multiplication

Polynomials can be multiplied approximately, which is useful for preventing rapid increase in the number of terms.


```python
print('Polynomial approximate multiplication example 1:')
pa = polynomialmath.Polynomial(np.array([1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2]))
pr_approx, prem = polynomialmath.multiply_polynomials_approx(pa, pb)
pr = pa * pb

print('({})  *  ({})'.format(pa.tostr(), pb.tostr()))
print('  = ({}) + ({})'.format(prem.tostr(), pr_approx.tostr()))
print('  = {}'.format(pr.tostr()))
print()

print('Polynomial approximate multiplication example 2:')
pa = polynomialmath.Polynomial(np.array([1, 1]))
pb = polynomialmath.Polynomial(np.array([2, 2]))
interval = intvmath.Interval(-1, 1)
pr_approx, prem, bounds = polynomialmath.multiply_polynomials_approx_and_bound(
    pa, pb, interval, n_subdomains=1)
pr = pa * pb

print('({})  *  ({})'.format(pa.tostr(), pb.tostr()))
print('  = ({}) + ({})'.format(prem.tostr(), pr_approx.tostr()))
print('  ~= {} + ({})'.format(bounds.tostr(), pr_approx.tostr()))
print('  = {}'.format(pr.tostr()))
print()
```

## Raising Polynomials to Integer Powers

A polynomial can be raised to an integer power.


```python
pa = polynomialmath.Polynomial(np.array([-1]))
inte = 2
pr = polynomialmath.pown_polynomial(pa, inte)

print()
print('({})^{} = {}'.format(pa.tostr(), inte, pr.tostr()))

pa = polynomialmath.Polynomial(np.array([-10.5, 3, -2]))
inte = 2
pr = polynomialmath.pown_polynomial(pa, inte)

print()
print('({})^{} = {}'.format(pa.tostr(), inte, pr.tostr()))

pa = polynomialmath.Polynomial(np.array([-3, 4, -5.5, -2]))
inte = 3
pr = polynomialmath.pown_polynomial(pa, inte)

print()
print('({})^{} = {}'.format(pa.tostr(), inte, pr.tostr()))

pa = polynomialmath.Polynomial(np.array([-3, 4, -5.5, -2]))
inte = 0
pr = polynomialmath.pown_polynomial(pa, inte)

print()
print('({})^{} = {}'.format(pa.tostr(), inte, pr.tostr()))
```

## Composing two polynomials

Two polynomials f(x) and g(x) can be composed to compute h(x) = f(g(x)).


```python
print('Polynomial composition example 1:')
pg = polynomialmath.Polynomial(np.array([1, 1, 1]))
pf = polynomialmath.Polynomial(np.array([1, 0, 0]))
pr = polynomialmath.compose_polynomials(pf, pg) 

print('f(x) = {}'.format(pf.tostr()))
print('g(x) = {}'.format(pg.tostr()))
print('f(g(x))  = {}'.format(pr.tostr()))
print()

print('Polynomial composition example 2:')
pg = polynomialmath.Polynomial(np.array([2, -1, 0.5]))
pf = polynomialmath.Polynomial(np.array([3, -1, 2, 10]))
pr = polynomialmath.compose_polynomials(pf, pg) 

print('f(x) = {}'.format(pf.tostr()))
print('g(x) = {}'.format(pg.tostr()))
print('f(g(x))  = {}'.format(pr.tostr()))
print()
```

## Evaluation

Polynomials can be evaluated over an array of points.


```python
pa = polynomialmath.Polynomial(np.array([-0.03, 0.3, 1, -3, 1]))
x = np.linspace(-5, 5, 100)
y = pa.evaluate(x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(
    xlabel='x', ylabel='y', title='y = {}'.format(pa.tostr())
)

pa = polynomialmath.Polynomial(np.array([1, 0, 0]))
x = np.linspace(-5, 5, 100)
y = pa.evaluate(x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(
    xlabel='x', ylabel='y', title='y = {}'.format(pa.tostr())
)
```

## Bounding Example 1

Bounds can be computed for a polynomial over a domain.  If the domain is small or if the function is monotonic, computing the bounds directly can be accurate.  Even for some simple functions that do not meet this criteria however, the bounds can be severaly overestimated (see below).


```python
pa = polynomialmath.Polynomial(np.array([0.3, 1, -3, 1]))
interval = intvmath.Interval(-1, 1)
bounds = pa.bound(interval)

print('Interval bounds on the domain:')
bounds.print()

x = np.linspace(interval.x1, interval.x2, 100)
y = pa.evaluate(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, np.ones(x.shape) * bounds.x1)
ax.plot(x, np.ones(x.shape) * bounds.x2)

ax.set(
    xlabel='x', ylabel='y', title='y = {}'.format(pa.tostr())
)
```

## Bounding Example 2: Taylor Series of Sine

Even for a simple function like sin(x) the bounds are severely overestimated!

```python
pa = polynomialmath.Polynomial(np.array([
    -1/39916800, 0, 1/362880, 0, -1/5040, 0, 1/120, 0, -1/6, 0, 1, 0]))
interval = intvmath.Interval(-np.pi-1, np.pi+1)
bounds = pa.bound(interval)

print('Interval bounds on the domain:')
bounds.print()

x = np.linspace(interval.x1, interval.x2, 100)
y = pa.evaluate(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, np.ones(x.shape) * bounds.x1)
ax.plot(x, np.ones(x.shape) * bounds.x2)

ax.set(
    xlabel='x', ylabel='y', title='Bounds')
```

## Bounding Example 3: Tight Bounds with Subdomains

Using subdomains fixes this problem by subdividing the interval of interest into many small subintervals which can be treated sepparately.


```python
pa = polynomialmath.Polynomial(np.array([
    -1/39916800, 0, 1/362880, 0, -1/5040, 0, 1/120, 0, -1/6, 0, 1, 0]))
interval = intvmath.Interval(-np.pi-1, np.pi+1)
bounds = pa.bound_with_subdomains(interval, n_subdomains=1000)

print('Interval bounds on the domain:')
bounds.print()

x = np.linspace(interval.x1, interval.x2, 100)
y = pa.evaluate(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, np.ones(x.shape) * bounds.x1)
ax.plot(x, np.ones(x.shape) * bounds.x2)

ax.set(
    xlabel='x', ylabel='y', title='Bounds')
```

## Example Application: Bounding sin(sin(x))

Here we combine several operations to compute bounds for sin(sin(x)).  Note that these bounds are not formalized since we are not using a Taylor model here.  Taylor models can be built from these primitives however.


```python
p_sin = polynomialmath.Polynomial(np.array([
    -1/39916800, 0, 1/362880, 0, -1/5040, 0, 1/120, 0, -1/6, 0, 1, 0], dtype=np.float64))

p_sinsin = polynomialmath.compose_polynomials(p_sin, p_sin)

interval = intvmath.Interval(-np.pi/2-0.2, np.pi/2+0.2)
bounds = p_sinsin.bound_with_subdomains(interval, n_subdomains=1000)

x = np.linspace(interval.x1, interval.x2, 100)
y_approx = p_sinsin.evaluate(x)
y_true = np.sin(np.sin(x))

fig, ax = plt.subplots()
ax.plot(x, y_true, linewidth=6)
ax.plot(x, y_approx)
ax.plot(x, np.ones(x.shape) * bounds.x1)
ax.plot(x, np.ones(x.shape) * bounds.x2)
ax.legend(('True', 'Polynomial', 'Upper Bound', 'Lower Bound'))

ax.set(
    xlabel='x', ylabel='y', title='Bounds')
```
