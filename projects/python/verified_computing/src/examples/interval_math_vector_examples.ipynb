## Interval Math Vector Examples

This notebook provides examples of how to use the intvmath library for some basic vector operations on intervals and vectors of intervals.


```python
%load_ext autoreload
%autoreload 2

import numpy as np

from verified_computing import intvmath
```

## Printing

Printing of interval vectors in pretty format is supported.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

print('Column vector of intervals:')
intv.print()
```

## Copying Interval Vectors

Interval vectors can be copied using different memory.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

intv_copy = intv.copy()
intv_copy.elems[0] = intvmath.Interval(-2, 2)

print('Original interval vector:')
intv.print()
print()

print('Interval vector copy:')
intv_copy.print()
```

## Addition

Interval vectors can be added.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

intv_r = intv + intv

print('Result:')
intv_r.print()
```

## Adding a Scalar to an Interval Column Vector

A scalar can be added to a column vector of intervals.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

scalar = 1.5

intv_r = intvmath.add_scalar_to_interval_column_vector(scalar, intv)

print('Result:')
intv_r.print()
```

## Adding an Interval Column Vector to a Real Vector

A vector of real numbers and an interval vector can be added.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

real_vector = np.array([-1.0, -1.0, -1.0])

intv_r = intvmath.add_real_vector_to_interval_column_vector(real_vector, intv)

print('Result:')
intv_r.print()
```

## Subtraction

Interval column vectors can be subtracted.  Note that subtracting intervals from themselves does not yeild zero.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

intv_r = intv - intv

print('Result:')
intv_r.print()
```

## Subtracting Intervals and Scalars

Intervals column vectors can be subtracted from scalars and scalars can be subtracted from interval column vectors.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(0, 1),
    intvmath.Interval(-1, 1)
])

scalar = 5

intv_r = intvmath.subtract_scalar_from_interval_column_vector(
    intv, scalar
)

print('Result:')
intv_r.print()
print()

intv_r = intvmath.subtract_interval_column_vector_from_scalar(
    scalar, intv
)

print('Result:')
intv_r.print()
```

## Multiply Interval Column Vectors

Interval column vectors can be multiplied.  Note that self-multiplication is not the same as squaring an interval.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(3, 4),
    intvmath.Interval(-2, 2)
])

intv_r = intv * intv

print('Result:')
intv_r.print()
```

## Multiply an Interval Column Vector by a Scalar

Interval column vectors can be multiplied by scalars.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(3, 4),
    intvmath.Interval(-2, 2)
])

scalar = 5

intv_r = intvmath.multiply_interval_column_vector_by_scalar(
    scalar, intv
)

print('Result:')
intv_r.print()
```

## Raising and Interval Column Vector to an Integer Power

Interval column vectors can be raised to integer powers.  Note that this is not the same as self-multiplication.



```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(3, 4),
    intvmath.Interval(-2, 2)
])

inte = 2

intv_r = intvmath.pown_interval_column_vector(intv, inte)

print('Result:')
intv_r.print()
print()

inte = 0

intv_r = intvmath.pown_interval_column_vector(intv, inte)

print('Result:')
intv_r.print()
print()
```

## Multiplying an Interval Column Vector by a Real Vector

Interval vectors can be dot product multiplied with real vectors.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(3, 4),
    intvmath.Interval(-2, 2)
])

vector = np.array([1, 2, 3])

intv_r = intvmath.multiply_interval_column_vector_by_real_vector(
    vector, intv
)

print('Result:')
intv_r.print()
```

## Summing an Interval Vector

The sum of a vector of intervals can be computed.


```python
intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(3, 4),
    intvmath.Interval(-2, 2)
])

intv_r = intvmath.sum_interval_column_vector(intv)

print('Result:')
intv_r.print()
```

## Left Multiplying an Interval Column Vector by a Matrix

Interval column vectors can be left multiplied by matricies.


```python
matrix = np.array(
    [[1.0, 1.0, 1.0],
     [2.0, 2.0, 2.0]]
)

intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(1, 2),
    intvmath.Interval(1, 2)
])

intv_r = intvmath.left_multiply_interval_column_vector_by_matrix(
    matrix, intv
)

print('Result:')
intv_r.print()
print()

matrix = np.array(
    [[1.0, 1.0, 1.0],]
)

intv = intvmath.IntervalColumnVector([
    intvmath.Interval(1, 2),
    intvmath.Interval(1, 2),
    intvmath.Interval(1, 2)
])

intv_r = intvmath.left_multiply_interval_column_vector_by_matrix(
    matrix, intv
)

print('Result:')
intv_r.print()
```
