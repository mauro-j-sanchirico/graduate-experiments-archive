# Taylor Factory Examples

This notebook provides examples of how to use the taylorfactory library to generate Taylor models.


```python
%load_ext autoreload
%autoreload 2

import numpy as np

from verified_computing import taylorfactory
```

## Taylor Model for Tanh via Taylor Series

Here we construct a Taylor model for tanh via Taylor series.


```python
bound = np.pi/2 - 0.1

t_tanh_taylor = taylorfactory.get_taylor_model(
    function_type='tanh_taylor',
    big_m=10,
    v_max=bound,
    dps=100,
    silent=True,
    n_brute_points=100,
    tol_final=1e-3)

t_tanh_taylor.plot(x_min=-bound, x_max=bound)

print('Min and max error:')
print(t_tanh_taylor.error.x1, t_tanh_taylor.error.x2)
```
