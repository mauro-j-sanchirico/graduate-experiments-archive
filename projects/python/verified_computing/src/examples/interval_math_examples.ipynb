# Interval Math Examples

This notebook provides examples of how to use the intvmath library.  The intvmath library is a custom developed library for performing basic interval arithmetic operations.  It implements a subset of the IEEE-17881 and is meant
to serve as a foundation for formal verifed computing with polynomials via
Taylor models and Taylor-like models.  This notebook shows how to use the basic interval arithmetic functions provided by intvmath.


```python
%load_ext autoreload
%autoreload 2

import numpy as np

from verified_computing import intvmath
```

## Examples of how to use the basic operations

Here several examples are provided showing how to use the basic operations.

Basic operations supported:
* Interval addition
* Interval and scalar addition
* Interval - scalar subtraction
* Scalar - interval subtraction
* Interval and scalar subtraction
* Interval subtraction from scalar
* Interval multiplication
* Interval multiplication by scalar
* Even integer powers of intervals
* Odd integer powers of intervals
* Integer powers of intervals


## Printing

Printing of intervals in a pretty format is supported.


```python
print('Interval printing example:')
interval = intvmath.Interval(-10.2, 12.5)
interval.print()
print()
```

## Copying an Interval

An interval can be copied using the .copy() function.

```python
interval = intvmath.Interval(-1, 2)
interval_copy = interval.copy()
interval_copy.print()
```

## Addition

Interval-to-Interval addition and Interval-to-Scalar addition is supported.


```python
print('Interval addition example:')
interval_a = intvmath.Interval(1.0, 2.0)
interval_b = intvmath.Interval(4.0, 5.0)
interval_r = interval_a + interval_b
print('{} + {} = {}'.format(
    interval_a.tostr(), interval_b.tostr(), interval_r.tostr()))
print()

print('Interval addition example:')
interval_a = intvmath.Interval(1.0, 2.0)
interval_b = intvmath.Interval(-5.0, -4.0)
interval_r = interval_a + interval_b
print('{} + {} = {}'.format(
    interval_a.tostr(), interval_b.tostr(), interval_r.tostr()))
print()

print('Interval and scalar addition example:')
interval = intvmath.Interval(1.0, 2.0)
scalar = 5.0
interval_r = intvmath.add_scalar_to_interval(scalar, interval)
print('{} + {} = {}'.format(
    scalar, interval.tostr(), interval_r.tostr()))
print()
```

## Subtraction

Interval-from-Interval, Scalar-from-Interval, and Interval-from-Scalar subtraction are supported.


```python
print('Interval subtraction example:')
interval_a = intvmath.Interval(5.0, 7.0)
interval_b = intvmath.Interval(1.0, 2.0)
interval_r = interval_a - interval_b
print('{} - {} = {}'.format(
    interval_a.tostr(), interval_b.tostr(), interval_r.tostr()))
print()

print('Interval - scalar subtraction example:')
interval = intvmath.Interval(3.0, 4.0)
scalar = 1.0
interval_r = intvmath.subtract_scalar_from_interval(interval, scalar)
print('{} - {} = {}'.format(
    interval.tostr(), scalar, interval_r.tostr()))
print()

print('Scalar - interval subtraction example:')
scalar = -2
interval = intvmath.Interval(-4, -3)
interval_r = intvmath.subtract_interval_from_scalar(scalar, interval)
print('{} - {} = {}'.format(
    scalar, interval.tostr(), interval_r.tostr()))
print()
```

## Multiplication

Interval-by-Interval and Interval-by-Scalar multiplication are supported.


```python
print('Interval multiplication example:')
interval_a = intvmath.Interval(1.99, 2.01)
interval_b = intvmath.Interval(1.95, 2.03)
interval_r = interval_a * interval_b
print('{} * {} = {}'.format(
    interval_a.tostr(), interval_b.tostr(), interval_r.tostr()))
print()

print('Interval x scalar multiplication example:')
scalar = -5.0
interval = intvmath.Interval(2.0, 4.0)
interval_r = intvmath.multiply_interval_by_scalar(scalar, interval)
print('{} * {} = {}'.format(
    scalar, interval.tostr(), interval_r.tostr()))
print()
```

## Exponentiation

Intervals can be raised to even integer powers, odd integer powers, or arbitrary integer powers.  Note that non-integer powers of intervals are not
supported directly although a verified result for a non-integer power could be obtained by reducing the operation to multiplications and summations via a Taylor series or other polynomial approximation with known error bounds.


```python
print('Even integer power example:')
interval = intvmath.Interval(-2, 2)
expn = 2
interval_r = intvmath.pown_even_interval(interval, expn)
print('{} ^ {} = {}'.format(
    interval.tostr(), expn, interval_r.tostr()))
print()

print('Odd integer power example:')
interval = intvmath.Interval(-2, 2)
expn = 3
interval_r = intvmath.pown_odd_interval(interval, expn)
print('{} ^ {} = {}'.format(
    interval.tostr(), expn, interval_r.tostr()))
print()

print('Integer power example #1:')
interval = intvmath.Interval(-2, 2)
expn = 2
interval_r = intvmath.pown_interval(interval, expn)
print('{} ^ {} = {}'.format(
    interval.tostr(), expn, interval_r.tostr()))
print()

print('Integer power example #2:')
interval = intvmath.Interval(-2, 2)
expn = 3
interval_r = intvmath.pown_interval(interval, expn)
print('{} ^ {} = {}'.format(
    interval.tostr(), expn, interval_r.tostr()))
print()
```

## Difference between self-multiplication and integer powers

Note that multiplying an interval by itself is not the same as squaring it.  This is because multiplying an interal by itself carries the uncertainty of both operands of the multiplication operation whereas squaring an interval reduces the uncertainty in the result by eliminating the possibility of a negative result for real intervals.


```python
interval_r = intvmath.Interval(-3, 2) * intvmath.Interval(-3, 2)
interval_r.print()

interval_r = intvmath.pown_interval(intvmath.Interval(-3,2), 2)
interval_r.print()
```

## Subdivision of an interval

An interval can be subdivided into smaller intervals of equal length.  This is useful for breaking up high uncertainty calculations into several lower uncertainty calculations.


```python
interval = intvmath.Interval(-4, 7)
subintervals = interval.subdivide(8)

for subinterval in subintervals:
    subinterval.print()
```

## Bounding multiple intervals

The union of multiple intervals can be bounded to find the smallest interval which includes all input intervals.


```python
interval1 = intvmath.Interval(-4, 7)
interval2 = intvmath.Interval(-5, 2)
interval3 = intvmath.Interval(2, 20)
interval4 = intvmath.Interval(20, 21)

intervals = [
    interval1, interval2, interval3, interval4
]

bound_interval = intvmath.bound_intervals(intervals)
bound_interval.print()
```
