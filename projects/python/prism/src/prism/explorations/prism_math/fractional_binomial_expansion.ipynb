```python
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt
```

```python
def compute_fractional_binomial_approx(x, y, alpha, big_k):

    summation = np.zeros(x.size)

    for k in range(0, big_k+1):
        numer = gamma(alpha + 1) * x ** k * y ** (alpha - k)
        denom = gamma(k + 1) * gamma(alpha - k + 1)
        term = numer / denom
        summation = summation + term

    return summation

def compute_fractional_binomial_true(x, y, alpha):
    return (x + y) ** alpha
```

```python
 dx = 0.1
 
 x = np.arange(-10, 10, dx)
 y = np.arange(-10, 10, dx)

 big_k = 10
 alpha = 0.5

 approx = compute_fractional_binomial_approx(x, y, alpha, big_k)
 truth = compute_fractional_binomial_true(x, y, alpha)

 error = truth - approx

 fig, axs = plt.subplots(2, 1, figsize=(6,4))
 axs[0].plot(x, truth, color='navy', linewidth=8)
 axs[0].plot(x, truth, color='limegreen', linewidth=2)
 axs[1].plot(x, error)
```
