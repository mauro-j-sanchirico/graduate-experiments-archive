```python
import numpy as np
import cupy as cp
import time
```

```python
# Numpy on CPU
s = time.time()
x_cpu = np.linspace(-10, 10, 100000000)
y_cpu = np.sin(x_cpu)
e = time.time()
print(e - s)
```

```python
# CuPy and GPU
s = time.time()
x_gpu = cp.linspace(-10, 10, 100000000)
y_gpu = cp.sin(x_gpu)
cp.cuda.Stream.null.synchronize()
e = time.time()
print(e - s)
```
