## Generate the Multinomial Tables

Here parameters for mulinomial expansions covered by the tables are defined and the tables are generated and saved.


```python
m_num_terms_min = 3
m_num_terms_max = 3

n_exponent_min = 0
n_exponent_max = 10
```

```python
mc_table = get_multinomial_coefs(
    m_num_terms_min=m_num_terms_min,
    m_num_terms_max=m_num_terms_max,
    n_exponent_min=n_exponent_min,
    n_exponent_max=n_exponent_max,
    load_file=None, save=True, silent=False,
    save_str='m_{}-{}_n_{}-{}'.format(
        m_num_terms_min, m_num_terms_max, n_exponent_min, n_exponent_max))
```
