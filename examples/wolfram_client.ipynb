```python
# Wolfram Language Imports
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr

session = WolframLanguageSession()

# Display Latex
from IPython.display import display, Math
```

```python
integrand = wlexpr("x^3 * Csch[a*x]")
intvar = wlexpr("x")
integral = session.evaluate(wl.Integrate(integrand, intvar))
tex_integral = session.evaluate(wl.ToString(wl.TeXForm(integral)))
display(Math(tex_integral))
```
